```python
import pandas as pd
```

```python
# Load the Excel file
df = pd.read_excel('AI_HUB_OpenSearch Company List.xlsx')
```

```python
# Extract company names from the Excel file
company_names = df['Company Name'].tolist()
```

```python
# Base URL template
base_url = "https://inxeption.com/directory/artificial-intelligence/{}/1"
```

```python
# Create a list to store the URLs
business_card_urls = []
```

```python
# Generate and print the URLs for each company
for company_name in company_names:
    business_card_url = base_url.format(company_name)
    #business_card_urls.append(business_card_url)
    print(business_card_url)
```

```python
business_strings = []
for index, row in df.iterrows():
    Company_name = row['Company Name']
    Business_card_url = row['Business Card URL']
    business_string = f'[{Company_name}]("{Business_card_url}")'
    business_strings.append(business_string)

# Add the product strings as a new column in the DataFrame
df['business_string'] = business_strings
```

```python
# Generate and add the business card URLs to the DataFrame
df['Business Card URL'] = [base_url.format(company_name) for company_name in df['Company Name']]

# Save the DataFrame to an Excel file
df.to_excel('Business_Card_URLs.xlsx', index=False)

print('Business card URLs have been generated and saved to Business_Card_URLs.xlsx')
```
